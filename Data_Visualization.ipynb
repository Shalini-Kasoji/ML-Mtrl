```python
#Bar charts
import matplotlib.pyplot as plt
%matplotlib inline
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
import numpy as np
ypos=np.arange(len(company))
plt.bar(ypos,revenue)
```

```python
plt.xticks(ypos,company)
plt.xlabel('Companies')
plt.ylabel('revenue(bin)')
plt.title('US Tech Stocks')
plt.bar(ypos,revenue,label='Revenue')
plt.legend()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

plt.xticks(ypos,company)
plt.title('US Tech Stocks')

plt.bar(ypos,revenue,label='Revenue')
plt.bar(ypos,profit,label='Profit')

plt.legend()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

xpos=np.arange(len(company))

plt.xticks(ypos,company)
plt.title('US Tech Stocks')

plt.bar(xpos-0.2,revenue,label='Revenue')
plt.bar(xpos+0.2,profit,label='Profit')

plt.legend()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

xpos=np.arange(len(company))

plt.xticks(ypos,company)
plt.title('US Tech Stocks')

plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')

plt.legend()
```

```python
#horizontal graph
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

plt.yticks(ypos,company)
plt.title('US Tech Stocks')

plt.barh(ypos-0.2,revenue,label='Revenue')
plt.barh(ypos+0.2,profit,label='Profit')

plt.legend() 
```

```python
#Pie chart
exp_labels=['Home Rent','Food','Phone/Internet Bill','Car','Other Utilities']
exp_vals=[1400,600,300,410,250]

plt.pie(exp_vals,labels=exp_labels)

```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%')
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',shadow=True)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',shadow=True,explode=[0,0.2,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',shadow=True,explode=[0,0.2,0,0,0],startangle=180)
plt.show()
```
